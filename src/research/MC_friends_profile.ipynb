````python
import os
import cv2
import numpy
import matplotlib.pyplot as plt
````

````python
def show_image(image):
    plt.imshow(image)
    plt.show()
````

````python
new_image = cv2.imread('../test/nf_mc_friend_avatars/frame_158.jpg')
g_new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2GRAY)
````

````python
ret, threshed_img = cv2.threshold(
    cv2.cvtColor(new_image, cv2.COLOR_RGB2GRAY),
    11, 55, cv2.THRESH_BINARY
)
````

````python
thresh = cv2.dilate(threshed_img, None, iterations=5)
````

````python
thresh = cv2.erode(threshed_img, None, iterations=5)
````

````python
show_image(thresh)
````

````python
contours, heir = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
````

````python
t_img = new_image.copy()
````

````python
for c in contours:
    x, y, w, h = cv2.boundingRect(c)
    cv2.rectangle(t_img, (x, y), (x+w, y+h), (0, 255, 0), thickness=1, lineType=8, shift=0)
````

````python
show_image(t_img)
````
